**We will use BeautifulSoup Library of Python in order to web scrap the table from the Wikipedia Link. A table with the postal codes of Canada will be printed.**


```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup=BeautifulSoup(source,'lxml')
print(soup.title)
from IPython.display import display_html
tab = str(soup.table)
display_html(tab,raw=True)
```

**The previous table is now converted to Pandas DF to clean and preprocess it.**

```python
dfs = pd.read_html(tab)
df=dfs[0]
df.head()
```

**We will not take into account those rows where Borough is "Not assigned". We will replace the name of these neighbourhoods with names found in Borough.**

**For simplicity and data cleaning, the neighbourhoods with the same Postal Code will be combined.**

```python
df1 = df[df.Borough != 'Not assigned']

df2 = df1.groupby(['Postal Code','Borough'], sort=False).agg(', '.join)
df2.reset_index(inplace=True)

df2['Neighbourhood'] = np.where(df2['Neighbourhood'] == 'Not assigned',df2['Borough'], df2['Neighbourhood'])

df2.rename(columns={'Postal Code':'Postcode'},inplace=True)

df2
```

```python
df2.shape
```

**We now define the Latitude and Longitude of the Postal Code groupings**

```python
lat_lon = pd.read_csv('https://cocl.us/Geospatial_data')
lat_lon.head()
```

```python
lat_lon.rename(columns={'Postal Code':'Postcode'},inplace=True)
df3 = pd.merge(df2,lat_lon,on='Postcode')
df3.head()
```

**A selection of those Boroughs that contain "York" has been chosen.**

```python
df4 = df3[df3['Borough'].str.contains('York',regex=False)]
df4
```

**Map representing York Boroughs**

**Maps might not be seen on GitHub**

```python
map_toronto = folium.Map(location=[43.651070,-79.347015],zoom_start=10)

for lat,lng,borough,neighbourhood in zip(df4['Latitude'],df4['Longitude'],df4['Borough'],df4['Neighbourhood']):
    label = '{}, {}'.format(neighbourhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
    [lat,lng],
    radius=5,
    popup=label,
    color='blue',
    fill=True,
    fill_color='#3186cc',
    fill_opacity=0.7,
    parse_html=False).add_to(map_toronto)
map_toronto
```
