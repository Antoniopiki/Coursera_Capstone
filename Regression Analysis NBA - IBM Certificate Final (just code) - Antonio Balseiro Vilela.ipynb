```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['PTS']])
train_y= np.asanyarray(train[['Wins']])
regr.fit (train_x, train_y)

print ('Coefficient: ', regr.coef_)
print ('Itercept: ', regr.intercept_)
```

```python
plt.scatter(train.PTS, train.Wins, color='green')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Points")
plt.ylabel("Wins")
```

```python
from sklearn.metrics import r2_score
test_x = np.asanyarray(test[['PTS']])
test_y = np.asanyarray(test[['Wins']])
test_y = regr.predict(test_x)

print("Mean absolute error : %.2f" % np.mean(np.absolute(test_y - test_y)))
print("Residual sum of squares (MSE) : %.2f" % np.mean((test_y - test_y) **2))
print("R2-score: %.2f" % r2_score(test_y, test_y) )      
```

```python
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['PTS', 'FGM', '3PM', 'DREB', 'AST', 'OREB']])
train_y= np.asanyarray(train[['Wins']])
regr.fit (train_x, train_y)

print ('Coefficient: ', regr.coef_)
print ('Intercept: ', regr.intercept_)
```

```python
y_hat = regr.predict(test[['PTS', 'FGM', '3PM', 'DREB', 'AST', 'OREB']])
x = np.asanyarray(test[['PTS', 'FGM', '3PM', 'DREB', 'AST', 'OREB']])
y = np.asanyarray(test[['Wins']])

print("Residual sum of squares (MSE) : %.2f" % np.mean((y_hat - test_y) **2))
print('Variance score: %.2f' % regr.score(x,y))
print("R2 score : %.2f" % r2_score(y,y))
```
